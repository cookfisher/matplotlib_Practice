```
import matplotlib.pyplot as plt
import numpy as np

x = [2618,2860,2010,2791,2212,2184,2344,2692,2206,2914,3034,4240,1400,2257]
y = [7.06, 6.62, 7.27, 7.00 , 7.69, 7.42, 6.41, 6.89, 8.28, 7.23, 4.90, 6.67, 6.71, 7.02]

# comput coefficient
u=v=s=t=0
for i in range(len(x)):
  u += x[i]*y[i]
  v += np.power(x[i],2)
t = np.mean(x)
s = t*sum(y)
B = (u-s)/(v-len(x)*np.power(t,2))
print('B = ',B)
A = np.mean(y)-B*t
print('A = ',A)
print('n = ',len(x))
Sxx = v - len(x)*np.power(t,2)
SSR =0
for i in range(len(x)):
  SSR += np.power(y[i]-A-B*x[i],2)
print("Sxx = ",Sxx)
print("SSR = ",SSR)
TS = np.power((len(x)-2)*Sxx/SSR,0.5)*np.abs(B)
print("TS = ",TS)
print("SSR/n-2 = ",SSR/(len(x)-2))
print("Xaverage = ",t)
X0 = 2500
Ta = 2.179
Error = np.power(1/len(x)+np.power(X0-t,2)/Sxx,0.5)*np.power(SSR/(len(x)-2),0.5)*Ta
left=A+B*X0-Error
right=A+B*X0+Error
print("A+B*X0 = ",A+B*X0," Error is ",Error)
print("100(1-a)% confidence is ",left,", ",right)
# Markers: https://matplotlib.org/api/markers_api.html

plt.scatter(x, y, color='g')
# plt.scatter(x2, y2, marker='v', color='r')
# plt.scatter(x2, y3, marker='^', color='m')
plt.xlabel("Cigarettes per Person")
plt.ylabel("Leukemia Deaths per Year per 100K people")
plt.title('Scatter Diagram')
plt.show()

x16=[]
y16=[]
j=0
for i in range(1300,4400,100):
  x16.append(i)
  y16.append(A+B*x16[j])
  j+=1
plt.plot(x16,y16,color='b')
plt.xlabel("Cigarettes per Person")
plt.ylabel("Leukemia Deaths per Year per 100K people")
plt.title('Regression Line')
plt.show()
```
